# Ollama
## 下载
- https://ollama.com/download
## 运行
1. 本地url: `http://localhost:11434/`
### 注意事项
1. Ollama默认安装在C盘，指定目录安装运行：`OllamaSetup.exe /DIR=F:\common_tools\Ollama`
2. 模型默认下载到C盘，在Ollama设置中指定安装盘符
3. 个人电脑不要下载参数>=32b的,但低于32b的幻觉严重
## 模型查找
- 地址：https://ollama.com/search
- 运行：`ollama run deepseek-r1:1.5b`

## ChatOllama API

```python
from langchain_ollama import ChatOllama

BASE_URL = 'http://localhost:11434/'

ollama_llm = ChatOllama(
    model="deepseek-r1:1.5b",
    base_url=BASE_URL,
)

print(ollama_llm.invoke('价绍一下你自己'))
```

## init_chat_model API

```python

from langchain.chat_models import init_chat_model

# 可以省略
BASE_URL = 'http://localhost:11434/'

model = init_chat_model(
    model="qwen-max",
    model_provider='ollama',
    base_url=BASE_URL,
)
res = model.invoke('写一个短剧分镜脚本，时长10s')
print(res.content)
```
