# invoke()
- 支持提示词文本输入
- 支持消息列表（记忆的基础）

# 消息类型
- system: 系统消息，结构: `{"role":"system", "content":"你是一个数据分析助手"}`, 对应类：`SystemMessage(content="")`
- user/Human: 用户提示词，结构: `{"role":"user", "content":"今天深圳天气怎么样？"}`，对应类：`UserMessage(content="")`, 推荐使用`role=user`
- assistant: 模型回复内容，结构: `{"role":"assistant", "content":"深圳今天晴天，温度28度"}`，对应类：`AssistantMesage(content="")`
- tool: 工具消息，结构: `{"role":"tool", "content":"深圳今天晴天，温度28度", "tool_call_id": "call_xxxxxx"}`，对应类：`ToolMessage(content="")`

## JSON消息列表invoke

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
from pathlib import Path
import os

load_dotenv(Path('../.env'))

model = init_chat_model(
    model="qwen-max",
    model_provider='openai',
    api_key= os.getenv('DASHSCOPE_API_KEY'),
    base_url= os.getenv('DASHSCOPE_BASE_URL'),
)

messages = [
    {"role": "system", "content": "你是一名资深数学老师"},
    {"role": "user", "content": "什么是傅里叶级数"}
]

resp = model.invoke(messages)
print(resp.content)
```

## 类消息列表invoke

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage
from dotenv import load_dotenv
from pathlib import Path
import os

load_dotenv(Path('../.env'))

model = init_chat_model(
    model="qwen-max",
    model_provider='openai',
    api_key= os.getenv('DASHSCOPE_API_KEY'),
    base_url= os.getenv('DASHSCOPE_BASE_URL'),
)

messages = [
    SystemMessage(content="你是一名大学计算机学院老师"),
    HumanMessage(content="一个优秀的计算机人员需要学哪些东西，列出需要学习的技术和知识"),
    AIMessage(content="一个优秀的计算机专业人员需要掌握广泛的技术和知识，这不仅包括编程语言、算法与数据结构等基础技能，也涵盖了一些高级技术领域。以下是一份较为全面的学习清单，但请注意，根据个人兴趣和发展方向的不同，重点学习的内容也会有所差异。"),
    HumanMessage(content="但是现在有很多程序员都是没有学计算机组成原理、操作系统这些基础知识的，你怎么评价")
]

resp = model.invoke(messages)
print(resp.content)
```
