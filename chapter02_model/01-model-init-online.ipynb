```python
# 通义非原生API的大模型需要通过ChatOpenAI调用

from pathlib import Path
import os

from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from pydantic import SecretStr

# 加载项目根目录 .env（不依赖 notebook cwd）
load_dotenv(Path(r"d:\ai\langchainv1.2-study\.env"), override=True)

DASHSCOPE_API_KEY = os.getenv("DASHSCOPE_API_KEY")
# OpenAI 兼容模式地址，例如：https://dashscope.aliyuncs.com/compatible-mode/v1
DASHSCOPE_BASE_URL = os.getenv("DASHSCOPE_BASE_URL")

# qwen3.7-plus 需走 OpenAI 兼容接口，不能用 ChatTongyi（它走 dashscope 原生 API）
# 原生 API 可用模型示例：qwen-plus / qwen-max / qwen-turbo
llm_tongyi = ChatOpenAI(
    model="qwen3.7-plus",
    base_url=DASHSCOPE_BASE_URL,
    api_key=DASHSCOPE_API_KEY, # type: ignore[arg-type]
)

print(llm_tongyi.invoke("请介绍一下自己"))
```
