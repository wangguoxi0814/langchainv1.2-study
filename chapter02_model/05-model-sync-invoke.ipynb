# 同步调用
- invoke(): 单次调用
- stream()：流式调用，返回一个迭代器
- batch()： 批量调用

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
import os
from pathlib import Path
from rich import print as rprint

# 从.env文件中加载环境变量
load_dotenv(Path('../.env'), override=True)
DASHSCOPE_MODEL_NAME = os.getenv('DASHSCOPE_MODEL_NAME')
DASHSCOPE_MODEL_PROVIDER = os.getenv("DASHSCOPE_MODEL_PROVIDER")
DASHSCOPE_API_KEY = os.getenv("DASHSCOPE_API_KEY")
DASHSCOPE_BASE_URL = os.getenv("DASHSCOPE_BASE_URL")

model = init_chat_model(
    model=DASHSCOPE_MODEL_NAME,
    model_provider=DASHSCOPE_MODEL_PROVIDER,
    api_key=DASHSCOPE_API_KEY,
    base_url=DASHSCOPE_BASE_URL
)
```

## 流式调用

```python
for chunk in model.stream("写一首七言律诗，总结大模型的发展"):
    print(chunk.text, end="", flush=True)  # 逐token输出
    # print(chunk.text, end="", flush=False)  # 逐token输出
```

## 批量调用
### batch
- 按任务提交顺序响应
- 返回迭代器

```python
messages = [
    "你好，你是谁？",
    "2 + 3 * 5 = ?",
    "中国首都在哪里？"
]
responses = model.batch(messages)
for response in responses:
    rprint(response)
```

### batch_as_completed
- 按任务完成顺序响应
- 响应会乱序，但每个元素都是一个元组，元组第一个元素为对应问题所在索引，可根据索引重新排序
- 返回生成器

```python
messages = [
    "你好，你是谁？",
    "2 + 3 * 5 = ?",
    "中国首都在哪里？"
]
responses = model.batch_as_completed(messages)
for response in responses:
    rprint(response)
```
